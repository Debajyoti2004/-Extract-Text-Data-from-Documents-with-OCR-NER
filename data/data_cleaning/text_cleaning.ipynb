```python
import pandas as pd
import numpy as np
import string
```

```python
with open(r"C:\Users\Debajyoti\OneDrive\Desktop\Project 2\csv_files\taggedBusinessCard.txt",mode='r',encoding='utf8',errors='ignore') as f:
    text = f.read()

text
```

```python
data = list(map(lambda x :x.split('\t'),text.split('\n')))

df = pd.DataFrame(data[1:],columns=data[0])
df.head()
```

```python
import re

whitespace = string.whitespace
punctuation = string.punctuation

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'\s+', '', text) 
    text = re.sub(r'[^\w]', '', text)  
    return str(text)
```

```python
df['text'] = df['text'].apply(clean_text)
cleanedData = df[df['text'].str.strip() != '']
cleanedData.dropna(inplace=True)
```
